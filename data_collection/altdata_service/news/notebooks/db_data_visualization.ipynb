# Visualization of ticker mentions over time

##### Load dependency libraries

```python
!pip install pandas
!pip install matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import data_collection.altdata_service.news.config.sqlalchemy_connector as dbconn
```

##### Load data from a database into pandas dataframe

```python
news_data_df = pd.read_sql_table('news', dbconn.conn)
```

##### Create a new dataframe that mentions specific yticker value

```python
YTICKER_VALUE = 'IRSH'
news_data_df = news_data_df.loc[news_data_df['yticker'] == YTICKER_VALUE]
news_data_df
```

##### Visualize data


```python
agg = news_data_df.groupby(['published']).agg('count')
plt.plot(agg.index, agg['yticker'])
```
